```python
df['datetime'] = pd.DatetimeIndex(df['datetime'])   #changing it to string
```

```python
hour = []
for timestamp in df['datetime']:
    hour = np.append(hour, timestamp.hour)
```

```python
hour[:10]
```

```python
df['datetime'] = hour
```

```python
df['season'].unique()       #unique shows what data is there
```

```python
mapper = {
    'Spring' : 0,
    'Summer' : 1,
    'Fall' : 2,
    'Winter' : 3
}
df['season'] = df['season'].map(mapper)
```

```python
df.sample(5)
```

```python
df['holiday'].unique()
```

```python
mapper = {
    'Yes' : 1,
    'No' : 0
}
df['holiday'] = df['holiday'].map(mapper)
```

```python
df.sample(5)
```

```python
df['weather'].unique()
```

```python
mapper = {
    'Clear' : 0,
    'Cloudy' : 1,
    'Light rain/snow' : 2
}
df['weather'] = df['weather'].map(mapper)
```

```python
df.sample(5)
```

```python
df['datetime'] = df['datetime'] / 23
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
to_scale = ['temp', 'humidity', 'windspeed']       #scaling is done to have x and y value same and helps for conversion

for each_feature in to_scale:
    scaler = StandardScaler()
    df[each_feature] = scaler.fit_transform(df[each_feature].values.reshape(-1, 1))
```
